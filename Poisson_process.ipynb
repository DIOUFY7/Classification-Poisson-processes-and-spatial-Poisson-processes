```python
import numpy as np
import matplotlib.pyplot as plt

def simulation_poisson(lambda_param, temps_total):
    """
    Simule un processus de Poisson.

    Args:
        lambda_param (float): Le taux moyen d'événements (lambda).
        temps_total (float): La durée totale de la simulation.

    Returns:
        tuple: Un tuple contenant les temps d'arrivée des événements et le nombre d'événements à chaque temps.
    """

    temps_arrivee = []
    temps_actuel = 0

    while temps_actuel < temps_total:
        # Générer un temps d'inter-arrivée à partir d'une loi exponentielle
        temps_inter_arrivee = np.random.exponential(1 / lambda_param)

        # Mettre à jour le temps actuel
        temps_actuel += temps_inter_arrivee

        # Enregistrer le temps d'arrivée si il est dans la période de temps de simulation
        if temps_actuel < temps_total:
            temps_arrivee.append(temps_actuel)

    # Créer un tableau de temps pour le graphique
    temps = np.arange(0, temps_total+0.05, 0.05)

    # Calculer le nombre d'événements à chaque temps
    nombre_evenements = [sum(arrivee <= t for arrivee in temps_arrivee) for t in temps]
    nombre_evenements_arrivee=[sum(arrivee < t for arrivee in temps_arrivee) for t in temps_arrivee]
    return temps_arrivee,nombre_evenements_arrivee, (temps, nombre_evenements)

# Paramètres de la simulation
lambda_param = 10  # Taux moyen de 2 événements par unité de temps
temps_total = 10  # Durée totale de la simulation

# Exécuter la simulation
temps_arrivee, nombre_evenements_arrivee,(temps, nombre_evenements) = simulation_poisson(lambda_param, temps_total)

# Afficher les résultats
print("Temps d'arrivée des événements :", temps_arrivee)

# Tracer le nombre d'événements en fonction du temps
plt.step(temps, nombre_evenements, where='post')
plt.xlabel("Temps")
plt.ylabel("Nombre d'événements")
plt.title("Simulation d'un processus de Poisson")
plt.show()
```

```python
import numpy as np

class GaussianNaiveBayes:
    """
    Implémente le Naïf Bayes Gaussien à partir de zéro.
    """

    def fit(self, X, y):
        """
        Entraîne le modèle Naïf Bayes Gaussien.

        Args:
            X (array-like): Les données d'entraînement, shape (n_samples, n_features).
            y (array-like): Les étiquettes cibles, shape (n_samples,).
        """
        self.classes = np.unique(y)
        self.mean_and_var = {}  # Stocke la moyenne et la variance pour chaque classe et caractéristique
        self.class_prior = {}    # Stocke les probabilités a priori des classes

        n_samples, n_features = X.shape

        for c in self.classes:
            # Filtrer les données pour la classe actuelle
            X_c = X[y == c]
            n_c = X_c.shape[0]

            # Calculer la moyenne et la variance pour chaque caractéristique
            mean = np.mean(X_c, axis=0)
            var = np.var(X_c, axis=0)

            self.mean_and_var[c] = {
                'mean': mean,
                'var': var
            }

            # Calculer la probabilité a priori de la classe
            self.class_prior[c] = n_c / n_samples

    def gaussian_pdf(self, x, mean, var):
        """
        Calcule la fonction de densité de probabilité (PDF) gaussienne.

        Args:
            x (float): La valeur pour laquelle calculer la PDF.
            mean (float): La moyenne de la distribution gaussienne.
            var (float): La variance de la distribution gaussienne.

        Returns:
            float: La valeur de la PDF gaussienne.
        """
        eps = 1e-6  # Constante pour éviter la division par zéro
        exponent = np.exp(-((x - mean) ** 2) / (2 * (var + eps)))
        return (1 / (np.sqrt(2 * np.pi * (var + eps)))) * exponent





    def predict(self, X):
        """
        Prédit les classes pour les échantillons donnés.

        Args:
            X (array-like): Les données de test, shape (n_samples, n_features).

        Returns:
            array-like: Les classes prédites, shape (n_samples,).
        """
        y_pred = [self._predict(np.array(X.iloc[x])) for x in range(X.shape[0])]
        return np.array(y_pred)

    def _predict(self, x):
        """
        Prédit la classe pour un seul échantillon.

        Args:
            x (array-like): Un seul échantillon, shape (n_features,).

        Returns:
            int: La classe prédite.
        """
        posteriors = {}

        for c in self.classes:
            # Initialiser la probabilité a posteriori avec la probabilité a priori de la classe
            posterior = self.class_prior[c]

            # Calculer la vraisemblance des caractéristiques étant donné la classe
            mean = self.mean_and_var[c]['mean']
            var = self.mean_and_var[c]['var']
            #print(x)
            for i in range(x.shape[0]):
                posterior *= self.gaussian_pdf(x[i], mean[i], var[i])

            posteriors[c] = posterior

        # Retourner la classe avec la probabilité a posteriori la plus élevée
        return max(posteriors, key=posteriors.get)

```

```python
def inter_times_jump(temps_list,N_list):
  inter_times=[temps_list[0]]
  inter_jumps = [0]
  s_inter=[temps_list[0]]
  s_0=temps_list[0]
  for i in range(1,len(temps_list)):
    d=temps_list[i]-temps_list[i-1]
    s_0=s_0+d
    s_inter.append(s_0)
    inter_times.append(d)
    inter_jumps.append(N_list[i]-N_list[i-1])
  return inter_times,inter_jumps,s_inter
```

```python
def pred_lambda_poisson(lambda_params,temps,n):

  if n >= temps*lambda_params.max():
    return lambda_params.max()
  elif n <= temps*lambda_params.min():
    return lambda_params.min()
  else :
    M=lambda_params[temps*lambda_params>=n].min()

    m=lambda_params[temps*lambda_params<=n].max()

    if poisson.pmf(n,m*temps) > poisson.pmf(n,M*temps):
      return m
    else :
      return M
```

```python
def pred_lambda_process(lambda_params,temps_list,N_list):
  pred_lambda_process=[]
  for i in range(len(temps_list)):
    pred_lambda_process.append(pred_lambda_poisson(lambda_params,temps_list[i],N_list[i]))
  return pred_lambda_process
```

```python
def calculate_product_with_poisson_pmf(y, x, t):
    """
    Calculates the value of the given product expression using scipy.stats.poisson.pmf.

    The expression is:
    product_{i=1 to n} [ (y * (t_i - t_{i-1}))^(x_i - x_{i-1}) / (x_i - x_{i-1})! * e^(-y * (t_i - t_{i-1})) ]

    Each term in the product is interpreted as the PMF of a Poisson distribution:
    P(k; mu) = mu^k * e^(-mu) / k!
    where k = x_i - x_{i-1} and mu = y * (t_i - t_{i-1}).

    Args:
        y (float): The independent variable.
        x (list or tuple): A list or tuple of x values [x_0, x_1, ..., x_n].
                            Must have at least two elements for the product to be non-empty.
        t (list or tuple): A list or tuple of t values [t_0, t_1, ..., t_n].
                            Must have the same length as x.

    Returns:
        float: The calculated value of the product.

    Raises:
        ValueError: If x and t have different lengths, or if the length of x (and t)
                    is less than 2.
        ArithmeticError: If (x_i - x_{i-1}) is negative (invalid for Poisson k),
                         or if y * (t_i - t_{i-1}) is negative (invalid for Poisson mu).
    """
    n = len(x) - 1

    if len(t) != len(x):
        raise ValueError("The lengths of x and t lists must be equal.")
    if n < 1: # The product starts from i=1, so we need at least x_0, x_1, t_0, t_1
        raise ValueError("The x and t lists must have at least 2 elements for the product to be defined.")

    product_result = 1.0

    for i in range(1, n + 1):
        delta_t = t[i] - t[i-1]
        k = x[i] - x[i-1] # This is the 'k' for poisson.pmf

        # For Poisson distribution, k must be a non-negative integer.
        if  k < 0:

            raise ArithmeticError(f"x_i - x_{i-1} must be a non-negative integer for Poisson PMF: got {k}")

        # For Poisson distribution, mu (rate parameter) must be positive.
        # It's usually non-negative in theory, but 0 rate means 0 probability for k > 0.
        mu = y * delta_t # This is the 'mu' for poisson.pmf
        if mu < 0:
            raise ArithmeticError(f"y * (t_i - t_{i-1}) (rate parameter 'mu') must be non-negative for Poisson PMF: got {mu}")
        if mu == 0 and k > 0:
            # If mu is 0, P(K=0; 0) = 1, and P(K=k; 0) = 0 for k > 0.
            # scipy.stats.poisson.pmf handles this correctly.
            pass

        # Calculate the PMF for the current term
        try:
            current_term_pmf = poisson.pmf(k, mu)
        except Exception as e:
            # Catch potential errors from scipy.stats.poisson.pmf for invalid inputs
            raise ArithmeticError(f"Error calculating Poisson PMF for k={k}, mu={mu}: {e}")

        product_result =np.prod([product_result, current_term_pmf],dtype=np.float128)

    return product_result
```

```python
def pred_lambda_process_inter_process(lambda_params,temps_list,N_list):
  if len(temps_list) == 0:
    N=0
    T=0
  else :
    N=N_list[-1]
    T=temps_list[-1]
  pred_lambda_process=[]
  # inter_times,inter_jumps,_=inter_times_jump(temps_list,N_list)
  # print(lambda_params.max())
  # print(lambda_params.min())
  if N >= T*lambda_params.max():

    return int(lambda_params.max())
  elif N <= T*lambda_params.min():
    return int(lambda_params.min())

  else :
    #print(lambda_params[T*lambda_params>N].max())
    M=lambda_params[T*lambda_params>=N].min()

    m=lambda_params[T*lambda_params<=N].max()
    # print(m,M)
    # print(N,T)
    # print((m**N)*np.exp(-m*T) ,(M**N)*np.exp(-M*T))
    a=calculate_product_with_poisson_pmf(m,N_list,temps_list)
    b=calculate_product_with_poisson_pmf(M,N_list,temps_list)
    # print(a,b)
    if  a> b :
      return int(m)
    else :
      return int(M)
```

```python
def pred_y_process_max(lambda_params,temps,x_data):
  y_pred=[]
  for i in range(len(x_data)):
    # Get the list of predicted lambda values for the current data point
    l_predictions = pred_lambda_process(lambda_params,temps,x_data.iloc[i])

    # Convert to numpy array for efficient frequency counting
    l_predictions_array = np.array(l_predictions)

    # Get unique predicted values and their counts
    unique_values, counts = np.unique(l_predictions_array, return_counts=True)

    # Find the index of the maximum count
    max_count_index = np.argmax(counts)

    # Get the unique value corresponding to the maximum count
    most_frequent_prediction = unique_values[max_count_index]

    y_pred.append(most_frequent_prediction)
  return y_pred
```

```python
def pred_y_process_mean_3_max(lambda_params,temps,x_data):
  y_pred=[]
  for i in range(len(x_data)):
    # Get the list of predicted lambda values for the current data point
    l_predictions = pred_lambda_process(lambda_params,temps,x_data.iloc[i])
     # Convert to numpy array for efficient frequency counting
    l_predictions_array = np.array(l_predictions)

    # Get unique predicted values and their counts
    unique_values, counts = np.unique(l_predictions_array, return_counts=True)

    #three max values
    if len(unique_values) < 3:
      # Find the index of the maximum count
      max_count_index = np.argmax(counts)

      # Get the unique value corresponding to the maximum count
      most_frequent_prediction = unique_values[max_count_index]

      y_pred.append(most_frequent_prediction)
    else :
      l_ar=counts.argsort()
      #print(unique_values)
      l_ar=l_ar[[-3,-2,-1]]
      unique_values=unique_values[l_ar]

      y_pred.append(round(unique_values.mean()))
  return y_pred
```

##Maximum Likelihood of λ

From observed counts: If you observe N events in a fixed time interval T, the MLE of λ is simply: $\hat{\lambda} = \dfrac{N}{T}$
This is intuitive: it's the total number of events divided by the total observation time.

```python
def MLE_pred(sauts,Total_times,max_label):
  return sauts.apply(lambda x: max_label if round(x/Total_times) > max_label else (1 if round(x/Total_times) <= 0 else round(x/Total_times))),round(sauts/Total_times)
```
